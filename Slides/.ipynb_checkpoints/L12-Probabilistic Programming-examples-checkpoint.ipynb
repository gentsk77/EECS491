### Probabilistic Programming 

Probabilistic programming are programming languages that facilitate model-based inference (for a good introduction see Chapter 1 in [An Introduction to Probabilistic Programming](https://arxiv.org/abs/1809.10756) by Meent et al).

These languages make it simple to specify probabilistic models.  Recall the coin-flip example.  This model would be specified as

$$
\begin{align}
\theta & \sim \textrm{Beta}(\alpha, \beta)\\
     y & \sim \textrm{Bernoulli}(\theta)
\end{align}
$$

Notice that the model is hierarchical.  The variable $y$ (the number of heads follows a Bernoulli distribution, which is defined by the parameter $\theta$.  This in turn is follows a Beta distribution.

This can also be written as a graphical model.

To see how it would be specified in a probabilistic programming language, we use `Turing.jl`. There are also numerous other options, e.g. `pgmpy`, `PyMC3`, `Stan`, to name just a few.

```julia
using TikzPictures, Distributions, Random, Turing, StatsPlots
```

```julia
using TikzPictures
g = TikzPicture(L"""
\tikzset{>=latex}
\matrix [row sep=5mm, column sep=2mm] {
\node (alpha) {$\alpha$}; & & \node (beta) {$\beta$}; \\
 & \node (theta) {$\theta$}; & \\
 & \node (y) {$y$}; & \\
};
\draw [->] (alpha) -- (theta);
\draw [->] (beta)  -- (theta);
\draw [->] (theta) -- (y);
""", options="thick, font=\\large")
```

Specifying this model is straight forward.

```julia
using Distributions, Random
```

```julia
p_true = 0.25 # true probability of heads
Ns = 0:50;    # observations iterate over this range

# Draw coin flips from Bernoulli distribution
#Random.seed!(2)
data = rand(Bernoulli(p_true), last(Ns))
```

```julia
# define model in Turing
using Turing

@model coinflip(y) = begin
    p ~ Beta(1, 1) # prior for probability of heads
    N = length(y)  # number of observations
    for n in 1:N
        y[n] ~ Bernoulli(p) # draw flips from Bernoulli distr
    end
end;
```

```julia
# settings for Hamiltonian Monte Carlo (HMC) sampler
iterations = 200
ϵ = 0.05
τ = 10

# perform sampling
chain = sample(coinflip(data), HMC(iterations, ϵ, τ));
```

```julia
using StatsPlots
plot(chain)
```

```julia
describe(chain)
```

The [MCMCChains.jl](https://github.com/TuringLang/MCMCChains.jl) package can generate several types of plots.  The default (used above) is:

```julia
plot(chain, seriestype = (:traceplot, :mixeddensity))
```

```julia
# Just the sameplot trace
plot(chain, seriestype = :traceplot)
```

```julia
# or shorthand = traceplot; and overlay the true value
traceplot(chain)
plot!([0, 200], [p_true, p_true], color="red", lw=2)
```

```julia
density(chain)
```

```julia
histogram(chain)
```

```julia
# running average of mean
meanplot(chain)
```

```julia
# make summary of sampling process for p
psummary = Chains(chain[:p])
#dump(psummary)
```
